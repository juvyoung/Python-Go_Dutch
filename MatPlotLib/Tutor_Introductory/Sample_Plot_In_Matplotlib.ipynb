> ## Line plot
> **Here's how to create a line plot with text labels using [plot()][1]

[1]:https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

```python
# Data for plotting

t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2*np.pi*t)

# Note that using plt.subplots below is equivalent to using
# fig = plt.figure() and then ax = fig.add_subplot(111)
fig, ax = plt.subplots()
ax.plot(t,s)

ax.set(xlabel = 'time(s)', ylabel='voltage(mv)',
       title = 'About as simple as it gets, folks')
ax.grid()

fig.savefig("sampleplot.png")
plt.show()
```

+ **plot() function**
    + Line Styles  
    '-'ï¼š solid  
    '--': dashed line  
    '-.': dashed dot  
    ':' : dotted line

```python
plt.plot([1,2,3],[1,4,9],'go-',label= 'line 1',linewidth=2 )
plt.plot([1,2,3],[2,5,6],'r*-.',label= 'line 2')
plt.grid(1)
plt.legend()
plt.show()
```

> ## Multiple subplots in one figure
> **Multiple axes(i.e. subplots) are created with the [subplot()][1] function

[1]:https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html#matplotlib.pyplot.subplot

```python
x1 = np.linspace(0,5)
x2 = np.linspace(0,2)
```

```python
y1 = np.cos(2*np.pi*x1) * np.exp(-x1)
y2 = np.sin(2*np.pi*x2)

plt.subplot(2,1,1) # 2 row, 1 colum
plt.plot(x1,y1, 'o-')
plt.title(' A table of 2 subplots ')
plt.ylabel(' Y axis')
plt.grid(1)

plt.subplot(2,1,2, facecolor = 'y') # yellow background
plt.plot(x2,y2,'r:')
plt.xlabel('time(s)')
plt.ylabel('Y axis')
plt.grid(1)

plt.show()
```

+ **subplot() function   
  subplot(rows, colums, index, *keyword args)** 
     + Keyword arguments  
       { facecolor }ï¼šone of {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w' }   
       { polar    }: 
       { projection}:

> ## Pie charts
> **Display images using [pie()][1] function**

[1]:https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pie.html#matplotlib.pyplot.pie

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwiseï¼›
labels = 'Frogs','Dogs','cats','mices'
sizes =[ 15, 30, 45, 10]
explode = (0, 0.1, 0, 0) # only 'explode' the 2nd slice

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode = explode, labels= labels, autopct='%1.1f%%', shadow = True, startangle = 90)
ax1.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle

plt.show()
```

> ## Legends
> **Here's how to create a plot with text labels using [legends()][1]
>  + legend()
>  + legend( labels )
>  + legend( handles, labels )

[1]:https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html#matplotlib.pyplot.legend

```python
a_ls = b_ls = np.arange(0, 3, .02)  # x-axis
c_ls = np.exp(a_ls) 
d_ls = c_ls[::-1]  # reverse
```

```python
# Create plots with pre-defined labels.

fig,ax = plt.subplots()
ax.plot( a_ls, c_ls, 'b' , label = ' Model Length ')
ax.plot( a_ls, d_ls, 'r:', label = ' Data Length ')          # multi-plot
ax.plot (a_ls, c_ls + d_ls, 'g-.', label = ' Total length ') # multi-plot

# @loc:      Location of the legend
# @shadow:   None or bool
# @fontsize: font size of the legend
legend = ax.legend( loc = 'best', shadow = True, fontsize = 'large')

#Put a nicer background color 

plt.show()
```
