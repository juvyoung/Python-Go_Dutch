```python
# Import 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for plotting
t = np.arange( 0.0, 2.0, 0.01 )
s = 1 + np.sin(2 * np.pi * t)

# Note that using plt.subplots below is equivalent to using
# fig = plt.figure() and then ax = fig.add_subplot(111)
fig, ax = plt.subplots()
ax.plot( t, s, 'g', label = 'drawing' )

ax.set(xlabel='time (s)', ylabel='voltage (mV)' ,
       title='s = 1+ sin(2pix)')
ax.grid(1)
ax.legend()
#fig.savefig("test.png")
plt.show()
```

plt.subplots() is a function that returns a **tuple** containing a _**figure**_ and _**axes**_ object(s). Thus when using   
fig, ax = plt.subplots() you unpack this tuple into the variables fig and ax.   
Having fig is useful if you want to change figure-level attributes or save the figure as an image file later with fig.savefig()
